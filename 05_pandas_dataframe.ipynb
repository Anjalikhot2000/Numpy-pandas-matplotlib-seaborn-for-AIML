# Filtering dataframe

```python
ipl.head(2)
```

```python
# find all the final winners
mask=ipl["MatchNumber"]=="Final"
new_df=ipl[mask]
new_df[["Season","WinningTeam"]]
```

```python
# how many super over finishes have occured
(ipl["SuperOver"]=="Y").sum()
```

```python
# how many matches has csk won in kolkata
ipl[(ipl["WinningTeam"]=="Chennai Super Kings") & (ipl["City"]=="Kolkata")].shape[0]
```

```python
# toss winner is match winner in percentage
((ipl[ipl["TossWinner"]==ipl["WinningTeam"]].shape[0])/(ipl.shape[0]))*100
```

```python
# movies with rating higher than 8 and votes>10000
movies[(movies["imdb_rating"]>8)&(movies["imdb_votes"]>10000)].shape[0]
```

```python
# Action movies with rating higher than 7.5
movies[(movies["imdb_rating"]>8) &(movies["genres"].str.contains("Action"))].shape[0]
```

```python
# write a function that can return the track record of 2 teams against each other
def track(t1, t2):
    matches = ipl[((ipl["Team1"] == t1) & (ipl["Team2"] == t2)) |
                  ((ipl["Team1"] == t2) & (ipl["Team2"] == t1))]

    team1_wins = matches[matches["WinningTeam"] == t1].shape[0]
    team2_wins = matches[matches["WinningTeam"] == t2].shape[0]

    return team1_wins, team2_wins

res=track("Chennai Super Kings","Kolkata Knight Riders")
print(res)
```
