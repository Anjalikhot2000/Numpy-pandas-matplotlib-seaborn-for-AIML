```python
# write a function that can return the track record of 2 teams against each other
def track(t1, t2):
    matches = ipl[((ipl["Team1"] == t1) & (ipl["Team2"] == t2)) |
                  ((ipl["Team1"] == t2) & (ipl["Team2"] == t1))]

    team1_wins = matches[matches["WinningTeam"] == t1].shape[0]
    team2_wins = matches[matches["WinningTeam"] == t2].shape[0]

    return team1_wins, team2_wins

res=track("Chennai Super Kings","Kolkata Knight Riders")
print(res)
```

```python
points_df = pd.DataFrame(
    {
        '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)

points_df
```

```python
def euclidean(row):
  pt_A = row['1st point']
  pt_B = row['2nd point']

  return ((pt_A[0] - pt_B[0])**2 + (pt_A[1] - pt_B[1])**2)**0.5
```

```python
points_df['distance'] = points_df.apply(euclidean,axis=1)
points_df
```
