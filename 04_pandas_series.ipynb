# What is Pandas

Pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool, built on top of the Python programming language.

# Pandas Series
A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.

```python
# Import Pandas
import numpy as np
import pandas as pd
```

```python
# Series from list
country=["India","Pakistan","USA","Sri Lanka"]
pd.Series(country)

```

```python
# integer
runs=[31,24,78,56]
pd.Series(runs)
```

```python
# Custom index
marks=[67,57,100]
sub=["Maths","English","Science"]
pd.Series(marks,index=sub)
```

```python
# setting a name
marks=pd.Series(marks,index=sub,name="Anjali's Marks")
marks
```

```python
# Series from dict

marks={
    "Maths":67,
    "English":57,
    "Science":100   
}
marks=pd.Series(marks,name="Anjali's Marks")
```

# Some important Series methods

```python
# astype
# astype is used to change the datatype of a series
import sys
sys.getsizeof(vk)
```

```python
sys.getsizeof(vk.astype('int16'))
```

```python
# between
# between is used to filter the values between a range
vk[vk.between(20,50)].size
```

```python
# clip
# clip is used to limit the values between a range
vk.clip(20,50)
```

```python
# drop_duplicates
# drop_duplicates is used to drop the duplicate values
temp=pd.Series([1,11,2,2,3,4])
temp.drop_duplicates()
```

```python
temp.duplicated().sum()
```

```python
vk.duplicated().sum()
```

```python
temp=pd.Series([1,2,3,4,np.nan,np.nan,8,9,10])
temp
```

```python
#is null
temp.size
```

```python
temp.count()
```

```python
temp.isnull().sum()
```

```python
#dropna
# dropna is used to drop the null values
temp.dropna()
```

```python
# fillna
# fillna is used to fill the null values with a specific value
temp.fillna(temp.mean())
```

```python
# isin
# isin is used to filter the values which are present in a list
vk[(vk==49) | (vk==99)]
```

```python
vk[vk.isin([49,99])]
```

```python
# apply
# apply is used to apply a function to all the values in a series
bw.apply(lambda a:a.upper())
```

```python
a.apply(lambda x:"good day" if x>a.mean()  else "bad day")
```

```python
a.mean()
```

```python
# View
# View is used to create a view of a series. It is a reference to the original series. Any changes made to the view will affect the original series.
s=[1,2,3,4,5,5]
i=[1,2,3,4,5,6]
a=pd.Series(s,index=i)
s_view=a.view()
```

```python
a[1]
```

```python
s_view[1]
```
